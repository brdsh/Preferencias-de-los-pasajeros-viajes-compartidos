# Prueba la hipótesis:
"La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare cambia los sábados lluviosos".
- Hipotesis nula:
  -  La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare no cambia los sábados lluviososs
- Hipotesis alternativa:
 - La duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare cambia los sábados lluviosos

```python
df3['start_ts'] = pd.to_datetime(df3['start_ts'], format='%Y-%m-%dT%H:%M:%S')
print(df3.info())
```

```python
print(df3.head())
```

```python
good_weather_conditions=df3[df3['weather_conditions']=='Good']['duration_seconds']
print(good_weather_conditions.head())
print('-------------------------------------------')
bad_weather_conditions=df3[df3['weather_conditions']=='Bad']['duration_seconds']
print(bad_weather_conditions.head())
```

```python
alpha = 0.05
results = st.ttest_ind(good_weather_conditions, bad_weather_conditions)
print('valor p: ', results.pvalue) # extraer el valor p

if results.pvalue < alpha: # comparar el valor p con el umbral
    print("Rechazamos la hipótesis nula")
else:
    print("No podemos rechazar la hipótesis nula")
```

- Para probar estas hipótesis, utilicé una prueba t de Student independiente. La razón detrás de elegir esta prueba es que estamos comparando las medias de dos grupos independientes de datos (duración de viajes en sábados lluviosos y no lluviosos). La prueba t de Student es adecuada para esta situación cuando tenemos muestras relativamente pequeñas y no conocemos la desviación estándar de las poblaciones.

- Utilicé un nivel de significancia (alpha) de 0.05, lo que significa que estamos dispuestos a aceptar un 5% de probabilidad de cometer un error de tipo I al rechazar incorrectamente la hipótesis nula. Este es un valor comúnmente aceptado en la literatura científica.

- Se comparó el valor p resultante con el nivel de significancia (alpha). Dado que el valor p (6.517970327099473e-12) es significativamente menor que el nivel de significancia (alpha = 0.05), se rechaza la hipótesis nula, lo que indica que hay suficiente evidencia para concluir que la duración promedio de los viajes desde el Loop hasta el Aeropuerto Internacional O'Hare cambia los sábados lluviosos.
